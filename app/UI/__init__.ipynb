```python
import tkinter as tk
from tkinter import ttk

class WellDesignedButtonApp(tk.Tk):
    def __init__(self):
        super().__init__()

        self.title("Well Designed Button")
        self.geometry("300x150")
        self.configure(bg="#f0f0f0")  # Set background color

        # Create a frame for widgets
        self.main_frame = ttk.Frame(self, padding=10)
        self.main_frame.pack(fill=tk.BOTH, expand=True)

        # Style for the button
        self.style = ttk.Style()
        self.style.configure("WellDesigned.TButton", background="#4CAF50", foreground="green", font=("Helvetica", 12), padding=10)
        self.style.configure("WellDesigned12.TButton", background="#4CAF50", foreground="green", font=("Helvetica", 12), padding=10)
        

        # Add a button
        self.button = ttk.Button(self.main_frame, text="Click Me", style="WellDesigned12.TButton", command=self.button_clicked)
        self.button.pack(pady=10)

        # Hover effect
        self.button.bind("<Enter>", self.on_enter)
        self.button.bind("<Leave>", self.on_leave)

    def button_clicked(self):
        print("Button clicked!")

    def on_enter(self, event):
        self.style.configure("WellDesigned12.TButton", background="#45a049")

    def on_leave(self, event):
        self.style.configure("WellDesigned12.TButton", background="#4CAF50")

if __name__ == "__main__":
    app = WellDesignedButtonApp()
    app.mainloop()
```

```python
from tkinter import *
from tkinter.ttk import *
from tkinter import filedialog






class Controller(tk.Tk):
    def __init__(self, *args, **kwargs):
        tk.Tk.__init__(self, *args, **kwargs)

        # Container to hold pages
        self.geometry("600x400")
        self.configure(bg="#f0f0f0")
        container = tk.Frame(self)
        container.pack(side="right", fill="both", expand=True)

        # Configure container to expand
        container.grid_rowconfigure(0, weight=1)  # Make row 0 expandable
        container.grid_columnconfigure(0, weight=1)  # Make column 0 expandable

        # Dictionary to hold pages
        self.frames = {}
        self.style = ttk.Style()
        self.style.configure("button.TButton", background="#4CAF50", foreground="green", font=("Helvetica", 12), padding=10)
        self.style.configure('title.TLabel', font=("Helvetica", 24), foreground="blue")
        
        # Create and store each page
        for F in (LoadingModelPage, ChoosingTypePage):
            page_name = F.__name__
            frame = F(parent=container, controller=self)
            self.frames[page_name] = frame
            frame.grid(row=0, column=0, sticky="nsew")

        # Show start page
        self.show_frame("LoadingModelPage")

    def show_frame(self, page_name):
        frame = self.frames[page_name]
        frame.tkraise()

    def load_file(self, page_name):
        file_path = filedialog.askopenfilename()
        if file_path:
            with open(file_path, 'r') as file:
                # Here you can read and process the contents of the file
                content = file.read()
                print("File loaded successfully.")
                print("Content:")
                print(content)
                popup = Toplevel(self)
                label = Label(popup, text="loaded successfully!", font=("Helvetica", 16))
                label.pack(expand=True, anchor="center")
                popup.geometry("250x120")
                self.show_frame(page_name)
                return content
        else:
            return None


class LoadingModelPage(Frame):
    def __init__(self, parent, controller):
        Frame.__init__(self, parent)
        self.controller = controller

        label = Label(self, text="GUI", style='title.TLabel')
        label.pack(pady=10, padx=10)

        button = Button(self, text="load a model ", style='button.TButton',
                           command=lambda: controller.load_file('ChoosingTypePage'))
        button.pack()

class ChoosingTypePage(Frame):
    def __init__(self, parent, controller):
        Frame.__init__(self, parent)
        self.controller = controller

        label = Label(self, text="Choosing a type", style='title.TLabel')
        label.pack()
        prev = Button(self, text="Go to Start Page", command=lambda: controller.show_frame('LoadingModelPage'))
        prev.pack(anchor='nw')
        label_choose_type = Label(self, text='Choose type:')
        label_choose_type.pack(side='left', pady=(0,52))
    
        self.options = ["Option 1", "Option 2", "Option 3", "Option 4", "Option 5"]
        self.combobox = Combobox(self, values=self.options, state="readonly")
        self.combobox.pack(side='left', pady=(0,52))
    
        self.combobox.bind("<<ComboboxSelected>>", self.handle_selection)

    def handle_selection(self, event):
        selected_value = self.combobox.get()
        print("Selected value:", selected_value)


if __name__ == "__main__":
    app = Controller()
    app.mainloop()

```
